```
#This code demonstrates:
#   reading a CSV file loctaed at URL via requests library
#   & displaying each line with a line number  
import requests

link = "https://raw.githubusercontent.com/gheniabla/datasets/master/fips.csv"
resp = requests.get(link)
content = resp.text
lines = content.splitlines()
line_num =1
for line in lines:
    print(F"{line_num} ===> {line}")
    line_num = line_num + 1

```
