```python
from io import BytesIO
from zipfile import ZipFile
from urllib.request import urlopen
# or: requests.get(url).content

url = "https://karpathy.ai/lexicap/data.zip"
resp = urlopen(url)
myzip = ZipFile(BytesIO(resp.read()))


"""
#lets' print the names of all files in the zip file
for filename in myzip.namelist():
    print(filename)
"""

#let's pick one file for now
filename = myzip.namelist()[1]

print(filename)

#lets open the file
myfile = myzip.open(filename)

dir(myfile)

#let's read all lines
lines = myfile.readlines()
  
count = 0
#let's print each line 
for line in lines:
    count += 1
    print("Line{}: {}".format(count, line))

```
